```python
outliers = {}
```

```python
cat_numerical_df = df_handled.select_dtypes(include=['number'])
display(cat_numerical_df)
```

```python
def check_outliers(column_name):
    Q1 = cat_numerical_df[column_name].quantile(0.25)
    Q3 = cat_numerical_df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outl = cat_numerical_df[(cat_numerical_df[column_name] < lower_bound) | (cat_numerical_df[column_name] > upper_bound)]
    if len(outl) > 0:
        outliers[column_name] = len(outl)
```

```python
for col_name in cat_numerical_df.columns:
    check_outliers(col_name)
print(outliers)
```
